#### PIPELINE DE DADOS
- **As etapas na construção de um pipeline de dados se baseiam em: DATA ENGINEERING, DATA PREPARATION E ANALYTICS** 

- **Para iniciar a construção do pipeline:** é interessante ter a localização das fontes de onde virão os dados;

- **Para a etapa de Data Engineering:** Coletar as fontes de dados (sejam elas 'n' formatos de arquivos); Agrupar o tipo de informação e unir toda a base dos dados; Realizar Data Quality, removendo dados inconsistentes, sejam eles duplicatas, dados faltantes e realizar agrupamento dos dados faltantes para que possam ser carregados nas tabelas.

- **Para a etapa de Data Preparation:** Quando os dados estiverem agrupados, menos poluídos e consistentes passarão por refinamentos para que possa ser aplicado regras de negócio/tratativas, catalogação dos dados e possível visualização. 

- **Na etapa de Analytics:** Serão gerados relatórios, insights, análise preditiva, aplicação de Machine Learning e possível Dashboard da operação

:-------------------------------- | :----------------------: |-------------:

### Data Quality

- Higienizar e homogenizar o formato da coluna `document`

```python
# Higienização do formato dos dados que tenham caracteres de separação
def corrige_caracter (valor):
  return valor.replace('.','').replace('-','').replace('/','').replace(',','')

# Aplicação da higienização na coluna 'document'
base_limpa['document'] = base_limpa['document'].apply(corrige_caracter)
```

- Detectar através da coluna `document` se o registro é de uma Pessoa Física ou Pessoa Jurídica, adicionando uma coluna com essa informação

```python
# Detecção se o registro é PF ou PJ
def valida_doc (doc):
  if len(str(doc)) == 11:
    status = "Pessoa Física"
  if len(str(doc)) == 14:
    status = "Pessoa Jurídica"
  return status

# Aplicação da validação e criação da coluna 'tipo_doc'
base_limpa['tipo_doc'] = base_limpa['document'].apply(valida_doc)
```

- Higienizar e homogenizar o formato da coluna `birthDate`

```python
# Listando os formatos dos BirthDate encontrados
tipos_datas = list()
for birthDate in list (base_limpa['birthDate'].drop_duplicates()):
  if len(birthDate) > 5:
    tipos_datas.append(birthDate)

tipos_datas
```

```python
# Tratar os tipos de Birthdate para o padrão
# Regular Expression
import re 

# Listagem das 8 RE encontradas em tipos_datas
list_regex_dt = {
  "dd/mm/yyyy" : r'\b([0-9]{1,9})/([0-9]{2})/([0-9]{4})\b',                     # 0 
  "dd-mm-yyyy" : r'\b([0-9]{1,9})-([0-9]{2})-([0-9]{4})\b',                     # 1 
  "dd/mmm/yyyy": r'\b([0-9]{1,9})/([a-zA-Z]{3})/([0-9]{4})\b',                  # 2 
  "dd-mmm-yyyy": r'\b([0-9]{1,9})-([a-zA-Z]{3})-([0-9]{4})\b',                  # 3 
  "mmm/dd/yyyy": r'\b([a-zA-Z]{3})/([0-9]{2})/([0-9]{4})\b',                    # 4 
  "mmm-dd-yyyy": r'\b([a-zA-Z]{3})-([0-9]{2})-([0-9]{4})\b',                    # 5
  "sss, mmm.dd.yyyy":r'\b([a-zA-Z]{3}), ([a-zA-Z]{3}).([0-9]{2}).([0-9]{4})\b', # 6 
  "yyyymmdd":r'([0-9]{4})([0-9]{2})([0-9]{2}),'                                 # 7 
}

# Listagem das chaves Mes e valor
meses = {
  'Jan':'01',
  'Feb':'02',
  'Mar':'03',
  'Apr':'04',
  'May':'05',
  'Jun':'06',
  'Jul':'07',
  'Aug':'08',
  'Sep':'09',
  'Oct':'10',
  'Nov':'11',
  'Dec':'12' 
}
# Retorna valor do mes
def qual_mes(valor):
  ms = 0
  for mes in meses.keys():
    if mes == valor:
      ms = (meses[mes])
  return ms

# Padroniza formato birthDate para dd-mm-aaaa
def format_data(padrao,dados):
  padroes = list(list_regex_dt.keys())
  data = ""
  if padrao == padroes[0]: #  "dd/mm/yyyy"
    dia, mes, ano = dados 
    data = "{}-{}-{}".format(dia,mes,ano)  
     
  if padrao == padroes[1]: #  "dd-mm-yyyy"
    dia, mes, ano = dados 
    data = "{}-{}-{}".format(dia,mes,ano)  

  if padrao == padroes[2]: #  "dd/mmm/yyyy"
    dia, mes, ano = dados 
    mes = qual_mes(mes)
    data = "{}-{}-{}".format(dia,mes,ano)  

  if padrao == padroes[3]: #  "dd-mmm-yyyy"
    dia, mes, ano = dados 
    mes = qual_mes(mes)
    data = "{}-{}-{}".format(dia,mes,ano)  

  if padrao == padroes[4]: #  "mmm/dd/yyyy"
    mes,dia, ano = dados 
    mes = qual_mes(mes)
    data = "{}-{}-{}".format(dia,mes,ano)  
  
  if padrao == padroes[5]: #  "mmm-dd-yyyy"
    mes,dia, ano = dados 
    mes = qual_mes(mes)
    data = "{}-{}-{}".format(dia,mes,ano)     

  if padrao == padroes[6]: #  "sss, mmm.dd.yyyy"
    semana, mes,dia, ano = dados 
    mes = qual_mes(mes)
    data = "{}-{}-{}".format(dia,mes,ano)     

  if padrao == padroes[7]: #  "yyyymmdd"
    ano,mes, dia  = dados 
    data = "{}-{}-{}".format(dia,mes,ano)       
  return data

# Lista e aplica regra da RE       
def checking_valid_dates(dt):
  result = {"padrao":0,"result":0}
  dt = str(dt)
  for regex_dt in list(list_regex_dt.keys()):
    # Método que aplica a regra
    r = re.findall(list_regex_dt[regex_dt], dt)
    if len(r) == 0:
      pass 
    else:
      result['result'] = list(r)
      result['padrao'] = regex_dt
  if len(result['result']) == 0:
     return dt
  else:
    return format_data(result['padrao'],list(result['result'][0]))
```

### Particionamento de dados tratados com as regras descritas em `DATA QUALITY`

- Particionar em arquivos PARQUET por estado (`state`)
- Particionar em arquivos CSV por ano/mes/dia de nascimento (`birthDate`)

```python
siglas = base_final[['state','document']].groupby(['state']).count().T.to_dict().keys()

for sigla in list (siglas):
  print(f'{sigla}')
  query = f"state == '{sigla}'"
  print(f'{query}')
  local = f"{sigla}.parquet"
  print(local)
  base_final.query(query).to_parquet(local)
```

```python
def colum_ano (valor):
  valor = str(valor)
  return valor.split("-")[2]

base_final ['ano']= base_final['birthDate'].apply(colum_ano)
```

```python
base_final[['birthDate','ano']].groupby(["ano"]).count()
```

```python
anos = base_final[['ano','birthDate']].groupby(['ano']).count().T.to_dict().keys()

for ano in list (anos):
  print(f'{ano}')
  query = f"ano == '{ano}'"
  print(f'{query}')
  local = f"{ano}.CSV"
  print(local)
  base_final.query(query).to_csv(local,index = False)
```
